```python
import pandas as pd
import holidays 
from datetime import date
```

### Let's get some weather data

```python
weather_data = pd.read_csv('http://www.frontierweather.com/historicaldataonly/KNYC_daily.txt')
```

```python
weather_data.head()
```

```python
weather_data.shape
```

```python
weather_data.Date = pd.to_datetime(weather_data.Date)
```

```python
weather_data.set_index('Date', inplace = True)
```

```python
weather_data.index
```

```python
weather_data = weather_data['2013-01-01':'2017-12-31']
```

```python
weather_data.head()
```

```python
weather_data.tail()
```

```python
weather_data.shape
```

```python
weather_data = weather_data.drop(['Site4', 'Source', 'HDDs', 'CDDs'], axis=1)
```

```python
weather_data.head()
```

```python
weather_data.to_csv('weather_data.csv')
```

### Is it a holiday or a noliday? 

```python
us_holidays = holidays.UnitedStates()
```

```python
holiday_list = []
for i in weather_data.index:
    holiday_list.append(i in us_holidays)
```

```python
holiday_list
```

```python
holiday_data = pd.DataFrame(data=holiday_list, index=weather_data.index, columns=['Holiday'])
```

```python
holiday_data.head()
```

```python
holiday_data.tail()
```

```python
holiday_data['2017-07-04':'2017-07-07']
```

```python
holiday_data['2017-10-31':'2017-11-10']
```

```python
holiday_data.to_csv('holiday_data.csv')
```
